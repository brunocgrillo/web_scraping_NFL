```python
from leagueleaders import LeagueLeaders
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
start_time = time.time()
category = 'passing'
start_season = 2013
df_decade = pd.DataFrame()

for x in range(10):
    df = LeagueLeaders(category, str(start_season)).get_data()
    df['season'] = start_season
    df_decade = pd.concat([df_decade, df[:10]])
    start_season += 1
    
end_time = time.time()
print(end_time - start_time)
```

```python
df_decade.info()
```

```python
df_decade = df_decade.reset_index().rename(columns={'index': 'SRank'})
df_decade['Player'] = df_decade['Player'].str.replace('+', '', regex=True).str.replace('*', '', regex=True)
df_decade['SRank'] += 1
```

```python
qb_appearances = df_decade['Player'].value_counts()
qb_appearances
```

```python
# Obter times de cada jogaador
player_teams = df_decade.groupby('Player')['Tm'].unique().reset_index(name='Teams')

# Calcular o número de times diferentes para cada jogador
player_teams['Unique_Teams'] = player_teams['Teams'].apply(len)

# Filtrar jogadores que estiveram em mais de um time diferente
multi_team_players = player_teams[player_teams['Unique_Teams'] > 1].sort_values('Unique_Teams', ascending=False)

multi_team_players
```
