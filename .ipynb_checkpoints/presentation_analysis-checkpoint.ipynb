## **Apresentação de Módulos de Web Scraping da NFL**

Neste notebook, meu objetivo é demonstrar as funcionalidades de cada módulo criado para web scraping. Realizei pequenos tratamentos e análises simples para destacar a capacidade desta ferramenta.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### ***LeagueLeaders***

```python
from leagueleaders import LeagueLeaders

#Definindo variáveis
category = 'passing'
start_season = 2020
num_seasons = 3

#Criando uma lista de dataframes contendo as informações das 3 últimas temporadas
df_list = []

for season in range(start_season, start_season + num_seasons):
    #Obtendo os dataframes com o módulo LeagueLeaders
    df = LeagueLeaders(category, season).get_dataframe()
    df['season'] = season
    df_list.append(df)

#Concatenando os dataframes
df_seasons = pd.concat(df_list, ignore_index=True)
df_seasons.head()
```

```python
#Definindo colunas para análise
columns = ['Player', 'Age', 'Cmp', 'Att', 'TD',
           'Int', 'Rate', 'Wins', 'Losses']

#Separando vitórias, derrotas e empates da coluna QBRec
df_seasons[['Wins', 'Losses', 'Draws']] = df_seasons['QBrec'].str.split('-', expand=True)

#Excluindo linhas com valores nulos
analysis_df = df_seasons[df_seasons['Pos'] == 'QB'][columns].dropna()

#Tratando nomes dos jogadores
analysis_df['Player'] = analysis_df['Player'].replace(['\*', '\+'], '', regex=True)

#Convertendo tipos de dados
for col in columns:
    try:
        analysis_df[col] = analysis_df[col].astype(int)
    except:
        try:
            analysis_df[col] = analysis_df[col].astype(float)
        except:
            pass

analysis_df.head()
```

***Análise de correlação e regressão linear***

```python
corr = analysis_df.corr()

sns.heatmap(data=corr, annot=True)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 16))

for y, col_name in enumerate(['Cmp', 'Att', 'Int', 'Rate', 'Wins', 'Losses']):
    row = y // 2
    col = y % 2
    sns.regplot(data=analysis_df, x='TD', y=col_name, ax=axs[row, col])

plt.show()
```

***LeagueLeaders2***

```python
from leagueleaders import LeagueLeaders2

#Definindo variáveis
category = 'passing'
start_season = 2012
num_seasons = 10

#Criando uma lista de dataframes contendo as informações das 10 últimas temporadas
df_list = []

for season in range(start_season, start_season + num_seasons):
    #Obtendo os dataframes com o módulo LeagueLeaders
    df = LeagueLeaders2(category, season).get_dataframe()
    df['season'] = season
    df_list.append(df[:10])

#Concatenando os dataframes
df_seasons = pd.concat(df_list, ignore_index=True)
```

```python
df_seasons.head()
```

***Análisando jogadores que estiveram no top 10 jogando por mais de um time***

```python
#Obtendo times de cada jogaador
player_teams = df_seasons.groupby('Player')['Tm'].unique().reset_index(name='Teams')

#Calculando o número de times diferentes para cada jogador
player_teams['Unique_Teams'] = player_teams['Teams'].apply(len)

#Filtrando jogadores que estiveram no top 10 jogando por multiplos times
multi_team_players = player_teams[player_teams['Unique_Teams'] > 1].sort_values('Unique_Teams', ascending=False)

multi_team_players
```

***Players***

```python
from players import Players

#Obtendo dados da carreira de Tom Brady
brady_stats = Players('Tom Brady').get_dataframe()

brady_stats.head()
```

```python
brady_stats.info()
```

```python
#Excluindo linhas de soma
brady_stats = brady_stats[:-3]

#Criando coluna com TDS acumulados
brady_stats['Sum_TDS'] = brady_stats['TD'].cumsum()

brady_stats.tail()
```

***Análise de touchdowns ao longo da carreira***

```python
plt.figure(figsize=(18,6))
sns.set_style('whitegrid')
ax = sns.barplot(data = brady_stats, x='Year',y='Sum_TDS', color='#1450A3')
plt.xlabel('Year')
plt.ylabel('Touchdowns')
plt.show()
```

***TeamGameLogs***

```python
from gamelogs import TeamGameLogs

#Obtendo dados da temporada em que o Seattle Seahawks foi campeão do Super Bowl
seattle_logs = TeamGameLogs('Seahawks', '2013').get_dataframe()

seattle_logs.info()
```

```python
#Definindo colunas
columns = ['Month', 'Day', 'W-L', 'HomeGame', 'PassY_Offense', 'RushY_Offense']

#Renomeando colunas fora do padrão
seattle_logs = seattle_logs.rename(columns={'Unnamed: 5_level_1': 'W-L', 'Unnamed: 8_level_1': 'HomeGame'})

#Ajustando valores da coluna HomeGame
seattle_logs['HomeGame'] = seattle_logs['HomeGame'].replace('@', 'N').replace(pd.NA, 'Y')

#Separando mês e ano da coluna Date
seattle_logs[['Month', 'Day']] = seattle_logs['Date'].str.split(' ', expand=True)

seattle_analysis = seattle_logs[columns]

seattle_analysis.dropna(inplace=True)

seattle_analysis
```

***Análise de vitórias e derrotas***

```python
#Realizando contagem de vitórias e derrotas por mês
wins_per_month = seattle_analysis.groupby('Month')['W-L'].value_counts().reset_index(name='Count')
team_record = wins_per_month.groupby('W-L')['Count'].sum().reset_index(name='Count')

print('Vitórias por mês')
#Loop para printar cada mês juntamente com a quantidade de vitórias
for index, row in wins_per_month[wins_per_month['W-L'] == 'W'].iterrows():
    print(f'{row["Month"]} - {row["Count"]}')

print('\nDerrotas por mês')
#Loop para printar cada mês juntamente com a quantidade de derrotas
for index, row in wins_per_month[wins_per_month['W-L'] == 'L'].iterrows():
    print(f'{row["Month"]} - {row["Count"]}')

#Printando o resultado final do time
print(f'\nO time terminou a temporada com {team_record[team_record["W-L"] == "W"]["Count"].values[0]} vitórias e {team_record[team_record["W-L"] == "L"]["Count"].values[0]} derrotas')
```

***Análise comparativa entre jardas passadas e jardas corridas***

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(data=seattle_analysis, x='Month', y='PassY_Offense', color='#7AC142', ci=None, label='Passing Yards')
sns.barplot(data=seattle_analysis, x='Month', y='RushY_Offense', color='#002A5C', ci=None, label='Rushing Yards')

plt.title('Desempenho Ofensivo de Seattle')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.legend()

plt.show()
```
